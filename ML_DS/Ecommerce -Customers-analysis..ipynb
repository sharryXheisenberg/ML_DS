# Spliting the dataset for Training and Testing


```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['Avg. Session Length',
 'Time on App','Time on Website',
 'Length of Membership', ]]
```

```python
Y = df['Yearly Amount Spent']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size= 0.3,random_state=10)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

# Training the model with linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
```

```python
ln.fit(X_train, Y_train)
```

```python
coeff = ln.coef_
```

```python
coeffdf = pd.DataFrame(coeff, X.columns,
 columns=["Co-Efficients"])
coeffdf
```

# Predictions

```python
prediction = ln.predict(X_test)
```

```python
prediction
```

```python
ln.score(X_test,Y_test)
```

# Relation between actual labels and and predicted labels

```python
sns.scatterplot(x=prediction, y=Y_test)
plt.xlabel("Predicted Values")
plt.title("Analyzing the Linear Model")
```

# Error Analysis

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
Absolute_error = mean_absolute_error(Y_test, prediction)
Absolute_error 
```

```python
Squared_error = mean_squared_error(Y_test, prediction)
Root_Squared = math.sqrt(Squared_error)
Root_Squared
```

# Residual Analysis


```python
residuals = Y_test - prediction
```

```python
sns.displot(residuals, bins=45, kde=True)
```

# Conclusion
Interpreting the analysis suggests that the most influential factor for clients is their length of membership,
rather than the time spent on the app or website. Notably, among the predictors (desktop vs. app), the app
has a significantly stronger impact. Surprisingly, the time spent on the desktop website shows no correlation
with the amount of money spent

# SVM (support vector machine)

```python
from sklearn.svm import SVR

svm_model = SVR()


svm_model.fit(X_train, Y_train)


r2_svm = svm_model.score(X_test, Y_test)
print("R^2 score for SVM:", r2_svm)
```

# Descision trees

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor()

dt_model.fit(X_train, Y_train)

r2_dt = dt_model.score(X_test, Y_test)
print("R^2 score for Decision Trees:", r2_dt)
```

# Random Forest


```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()

rf_model.fit(X_train, Y_train)

r2_rf = rf_model.score(X_test, Y_test)
print("R^2 score for Random Forest:", r2_rf)
```

```python
import matplotlib.pyplot as plt

r2_scores = [0.9871395472085015, 0.8582663372262341, 0.9269296719997961]

algorithms = ['Linear Regression', 'Decision Trees', 'Random Forest']
plt.figure(figsize=(10, 6))
plt.bar(algorithms, r2_scores, color='skyblue')
plt.xlabel('Algorithms')
plt.ylabel('R^2 Score')
plt.title('Comparison of R^2 Scores for Different Algorithms')
plt.ylim(0, 1)  
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
